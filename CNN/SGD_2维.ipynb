# 二维: pred = w1x1  + w2x2 + b,  y = 3.5x1 + (-1.5)x2 + 4

# 线性回归 loss functino: loss = (pred-target)^2

```python
import pandas as pd
import random
import math
import numpy as np

xs = []
targets = []

# generate x and label y
for i in range (100):
    for j in range(100):
        i = float(i)
        j = float(j)
        xs.append((i,j))
        targets.append(3.5 * i + (-1.5) * j + 4)
```

```python
# 观察数据
pd.DataFrame({
    'xs':xs,
    'targets':targets
})
```

```python
# randomly generate w and b, then use SGD to gradually fit
w1 = random.random() # random.random: generate value between [0.0, 1.0]
w2 = random.random()
b = random.random()

train_tuple = list(zip(xs, targets))
```

```python
learning_rate = 0.00001
log = []

for batch_number in range(500):  # i.e. run 500 times against all traing set
    (x1, x2), target = random.choice(train_tuple)  # 随机抽取一个<x,y> i.e.相当于batch_size =1
    pred = w1 * x1 + w2 * x2 + b
    delta_w1 = 2 * (pred - target) * x1  # 对loss求偏w的导数. loss = (pred-target)^2
    delta_w2 = 2 * (pred - target) * x2  #
    delta_b = 2 * (pred - target)   # 对loss求偏b的导数. loss = (pred-target)^2
    
    w1 = w1 - learning_rate * delta_w1
    w2 = w2 - learning_rate * delta_w2
    b = b - learning_rate * delta_b
    
    loss = pow (w1 * x1 + w2 * x2 + b - target, 2)
    
    log.append([w1, w2, b, loss])
    
```
