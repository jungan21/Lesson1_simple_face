# 一维: pred = wx  + b,  y = 3.5x + 4

# 线性回归 loss functino: loss = (pred-target)^2

```python
import pandas as pd
import random
import math
import numpy as np

xs = []
targets = []
# generate x and label y
for i in range (100):
    i = float(i)
    xs.append(i)
    targets.append(3.5 * i + 4)
```

```python
# 观察数据
pd.DataFrame({
    'xs':xs,
    'targets':targets
})
```

```python
# randomly generate w and b, then use SGD to gradually fit
w = random.random() # random.random: generate value between [0.0, 1.0]
b = random.random()

train_tuple = list(zip(xs, targets))
```

```python
learning_rate = 1e-5
log = []

for batch_number in range(500):  # i.e. run 500 times against all traing set
    x, target = random.choice(train_tuple)  # 随机抽取一个<x,y> i.e.相当于batch_size =1
    pred = w * x + b
    delta_w = 2 * (pred - target) * x  # 对loss求偏w的导数. loss = (pred-target)^2
    delta_b = 2 * (pred - target)   # 对loss求偏b的导数. loss = (pred-target)^2
    
    w = w - learning_rate * delta_w
    b = b - learning_rate * delta_b
    
    loss = pow (w * x + b - target, 2)
    
    log.append([w, b, loss])
    
```
